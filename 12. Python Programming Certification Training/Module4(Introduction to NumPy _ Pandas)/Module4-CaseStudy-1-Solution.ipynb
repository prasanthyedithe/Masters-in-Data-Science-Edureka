#### 5. How do you Count The Number Of Times Each Value Appears In An Array Of Integers?
[0, 5, 4, 0, 4, 4, 3, 0, 0, 5, 2, 1, 1, 9]
Answer should be array([4, 2, 1, 1, 3, 2, 0, 0, 0, 1]) which means 0 comes 4 times, 1 comes 2 times, 2 comes 1 time, 3 comes 1 time and so on.

```python
array = [0, 5, 4, 0, 4, 4, 3, 0, 0, 5, 2, 1, 1, 9]

count_array_elements = [array.count(a) for a in set(array)]

print(count_array_elements)
```

#### 7. Create a numpy array having NaN (Not a Number) and print it.
array([ nan, 1., 2., nan, 3., 4., 5.])
Print the same array omitting all elements which are nan

```python
import numpy as np

np_array = np.array([np.NaN,   1.,   2.,  np.NaN,   3.,   4.,   5.])
print(np_array)

arr_without_nan = np_array[~np.isnan(np_array)]  # use ~ for negation
print(arr_without_nan)
```

#### 14. Create a random matrix and Compute a matrix rank.

```python
import numpy as np

np_array = np.random.random(16).reshape(2, 2, 2, 2)
print(np_array)

array = np.array([4, 2, 7, 1])
temp = array.argsort()
ranks = np.empty_like(temp)
ranks[temp] = np.arange(len(array))

print(ranks)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

# Phase 1
df_school_data = pd.read_csv("middle_tn_schools.csv")
print(df_school_data.head())
df_school_data.describe()

# Phase 2
df_grouped_data = df_school_data.groupby("school_rating").describe()
print(df_grouped_data.head())

# Phase 3
corrmat = df_grouped_data.corr()
print(corrmat)
f, ax = plt.subplots(figsize=(9, 8))
sns.heatmap(corrmat, ax=ax, cmap="YlGnBu", linewidths=0.1)

# Phase 4 
plt.scatter(df_school_data["school_rating"], df_school_data["reduced_lunch"])
plt.grid()
plt.xlabel("Rating")
plt.ylabel("Reduced lunch")
plt.title("School rating vs Reduced lunch")
plt.show()
```
