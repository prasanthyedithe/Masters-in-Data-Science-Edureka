```python
import folium
from folium import plugins
import pandas as pd

print(folium.__version__)

pdiy_2016 = pd.read_csv("Police_Department_Incidents_Year_2016_.csv")


CHICAGO_COORD = [37.50, -122.403]

map_heat = folium.Map(CHICAGO_COORD,
               zoom_start=11)

for index, row in pdiy_2016.head(100).iterrows():
  folium.CircleMarker([row['X'], row['Y']],
                        radius=15,
                        popup=row['Category'],
                        fill_color="#3db7e4",  # divvy color
                        ).add_to(map_heat)

stationArr = [[row['X'],row['Y']] for index, row in pdiy_2016.iterrows()]

# plot heatmap
map_heat.add_child(plugins.HeatMap(stationArr, radius=15))
map_heat.save("BikeStationHeatMap.html")

```
