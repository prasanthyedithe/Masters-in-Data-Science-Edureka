### 1. Write a program which will find factors of given number and find whether the factor is even or odd.


```python
factors_list = []
even_list = []
odd_list = []

def factors(x):
   for i in range(1, x + 1):
       if x % i == 0:           
           factors_list.append(i)

numparameter = input("Enter a Number: ")
factors(int(numparameter))
print("The factors of",numparameter,"are:",factors_list)

for num in factors_list:
  if(num % 2 == 0):
    even_list.append(num)
  else:
    odd_list.append(num)

print("List of Even Numbers:",even_list)
print("List of Odd Numbers:",odd_list)
```

#### 2. Write a code which accepts a sequence of words as input and prints the words in a sequence after sorting them alphabetically. 

```python
my_str = input("Enter a string: ")  
# breakdown the string into a list of words  
words = my_str.split()  
# sort the list  
words.sort()  
# display the sorted words  
for word in words:  
   print(word) 
```

#### 4. Write a program that accepts a sentence and calculate the number of letters and digits.


```python
def countNumberString(inputParm):
  count_int = 0
  count_str = 0
  my_param_list = [x for x in inputParm]
  print(my_param_list)
  for val in my_param_list:
    if val.isdigit():
      count_int += 1
    elif isinstance(val, str):
      count_str += 1
  
  print ("LETTERS:",count_str)
  print("DIGITS:",count_int)


input_parm = input("Enter the value : ")
countNumberString(input_parm)

```

#### 5. Design a code which will find the given number is Palindrome number or not

```python
num=int(input("Enter a number:"))
temp=num
rev=0
while(num>0):
    dig=num%10
    rev=rev*10+dig
    num=num//10
if(temp==rev):
    print(temp, ":The number is palindrome!")
else:
    print(temp, ":Not a palindrome!")
```
