#### 3.A website requires a user to input username and password to register. Write a program to check the validity of password given by user. Following are the criteria for checking password:
<br>1. At least 1 letter between [a-z]
<br>2. At least 1 letter between [A-Z]
<br>3. At least 1 number between [0-9]
<br>4. At least 1 character from [$#@]
<br>5. Minimum length of transaction password: 6
<br>6. Maximum length of transaction password: 12

```python
def password_check(passwd):
    SpecialSym =['$', '@', '#']
    val = True

    if len(passwd) < 6:
        print('length should be at least 6')
        val = False
          
    if len(passwd) > 12:
        print('length should be not be greater than 12')
        val = False
          
    if not any(char.isdigit() for char in passwd):
        print('Password should have at least one numeral')
        val = False
          
    if not any(char.isupper() for char in passwd):
        print('Password should have at least one uppercase letter')
        val = False
          
    if not any(char.islower() for char in passwd):
        print('Password should have at least one lowercase letter')
        val = False
          
    if not any(char in SpecialSym for char in passwd):
        print('Password should have at least one of the symbols $@#')
        val = False
    if val:
        return val



# Main method
def main():
    passwd = input("Enter Password :")
      
    if (password_check(passwd)):
        print("Password is valid")
    else:
        print("Invalid Password !!")
          
# Driver Code        
if __name__ == '__main__':
    main()
```

#### 5.Please   write   a   program   which accepts  a   string   from   console   and   print   the characters that have even indexes.

```python

def printEvenIndexes(inputParm):
  final_str = ""
  my_param_list = [x for x in inputParm]
  for value in range(len(my_param_list)):
    if(value % 2 == 0):
      final_str = final_str + my_param_list[value]
  
  print(final_str)

printEvenIndexes("H1e2l3l4o5w6o7r8l9d")
```

#### 6.Please write a program which accepts a string from console and print it in reverse order.

```python
word = input("Input a word to reverse: ")

for char in range(len(word) - 1, -1, -1):
  print(word[char], end="")
print("\n")
```

#### Please write a program which count and print the numbers of each character in a string input by console.

```python
string = input("Enter a string: ")
lst1 = []
for char in string:
    if char not in lst1:
        lst1.append(char)
for item in lst1:
    print(item,string.count(item), sep = ",")
```

#### 8.With   two   given   lists   [1,3,6,78,35,55]   and   [12,24,35,24,88,120,155],   write   a program to make a list whose elements are intersection of the above given lists.

```python
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
 
# Driver Code
lst1 =[1,3,6,78,35,55]
lst2 = [12,24,35,24,88,120,155]
print(intersection(lst1, lst2))
```

#### 11.By using list comprehension, please write a program to print the list after removing the 0th,4th,5th numbers in [12,24,35,70,88,120,155].

```python

# initializing list
test_list = [12,24,35,70,88,120,155]
 
# printing original list
print("The original list is : " + str(test_list))
 
# initializing idx list
idx_list = [0, 4, 5]
 
# one-liner to test for element in index list
res = [ele for idx, ele in enumerate(test_list) if idx not in idx_list]
 
# printing results
print("Filtered List after removal : " + str(res))
```

#### 12.By using list comprehension, please write a program to print the list after removing delete numbers which are divisible by 5 and 7 in [12,24,35,70,88,120,155].

```python
input_lst = [12,24,35,70,88,120,155]

output = [num for num in input_lst if (num % 5 == 0 or num % 7 == 0)]

final_list = [value for value in input_lst if value not in output]

print(final_list)
```
